## 将6号项目3.12的数据和2.14原6号项目转出至11号项目的部分进行合并

```python
df_in_final = pd.concat([df6_0312_in,df6_0213_in],axis=0,join='outer',ignore_index=True,sort=False) 
#合并6号项目2月13日和3月12日两张数据库
```

```python
df_in_final['还款日期'] = pd.to_datetime(df_in_final['还款日期'],format='%Y%m%d')  #转换成日期格式
```

```python
df_in_final
```

### 添加描述列

#### 定义函数

```python
def describe(contract):
    if contract in(from11_0214):
        return '11号2.14转入'
    elif contract in(from1_0214keep):
        return '1号转入,2.14保留'
    elif contract in(from0_0214keep):
        return '0号转入,2.14保留'
    elif contract in(from0_0214out):
        return '0号转入,2.14转出'
    elif contract in (from1_0214out):
        return '1号转入,2.14转出'
    else:
        return '6号'
```

```python
df_in_final['Description'] = df_in_final['合同号'].apply(lambda x:describe(x))  #对6号项目3.12各个合同号的来源进行描述性匹配
```

```python
df_in_final['Description'].isna().value_counts()  #检查是否存在未匹配情况：无
```

#### 接下来，我们根据转入项目的时间，需要删除6号项目收款记录中不属于转入后发生的收款

```python
# 0号项目删除条件：还款日期在2019年9月25日之前
index_0 = df_in_final[(df_in_final['Description'].str.contains('0号转入')) & (df_in_final['还款日期']<'2019-09-25')].index
```

```python
# 1号项目删除条件：还款日期在2019年9月25日之前
index_1 = df_in_final[(df_in_final['Description'].str.contains('1号转入')) & (df_in_final['还款日期']<'2019-09-25')].index
```

```python
# 11号项目删除条件：还款日期在2020年2月15日之前
index_11 = df_in_final[(df_in_final['Description'].str.contains('11号')) & (df_in_final['还款日期']<'2020-02-15')].index
```

```python
# 汇总待删除的数据行编码
index_delete = index_0.append([index_1,index_11])
```
